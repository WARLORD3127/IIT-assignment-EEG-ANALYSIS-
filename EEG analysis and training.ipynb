```python
import mne
import numpy as np
import matplotlib.pyplot as plt
from mne.time_frequency import psd_array_welch

# Define parameters
fmin = 1  # Minimum frequency of interest
fmax = 100  # Maximum frequency of interest

# Define frequency bands
freq_bands = {'Delta': [1, 4],
              'Theta': [4, 8],
              'Alpha': [8, 12],
              'Beta': [12, 30],
              'Gamma': [30, 100]}

# Initialize lists to store PSD data
psd_rest = []
psd_task = []

# Iterate over each subject's data
for subject_id in range(36):
    file_rest = f"D:\\IIT Roorkhe assignment\\dataset\\eeg-during-mental-arithmetic-tasks-1.0.0\\eeg-during-mental-arithmetic-tasks-1.0.0\\Subject{subject_id:02d}_1.edf"
    file_task = f"D:\\IIT Roorkhe assignment\\dataset\\eeg-during-mental-arithmetic-tasks-1.0.0\\eeg-during-mental-arithmetic-tasks-1.0.0\\Subject{subject_id:02d}_2.edf"
    
    # Load data
    raw_rest = mne.io.read_raw_edf(file_rest, preload=True)
    raw_task = mne.io.read_raw_edf(file_task, preload=True)
    
    # Apply band-pass filter
    raw_rest.filter(l_freq=1, h_freq=100)
    raw_task.filter(l_freq=1, h_freq=100)
    
  # Calculate PSD for rest state
    psd_rest_subject, freqs_rest = psd_array_welch(raw_rest.get_data(), sfreq=raw_rest.info['sfreq'], fmin=fmin, fmax=fmax)
    psd_rest.append(psd_rest_subject)

    # Calculate PSD for task state
    psd_task_subject, freqs_task = psd_array_welch(raw_task.get_data(), sfreq=raw_task.info['sfreq'], fmin=fmin, fmax=fmax)
    psd_task.append(psd_task_subject)

# Convert lists to numpy arrays for easier manipulation
psd_rest = np.array(psd_rest)
psd_task = np.array(psd_task)

# Calculate average PSD across segments for each subject
avg_psd_rest = np.mean(psd_rest, axis=0)
avg_psd_task = np.mean(psd_task, axis=0)

# Plot PSD for each frequency band
for band, (fmin, fmax) in freq_bands.items():
    idx_band = np.where((freqs_rest >= fmin) & (freqs_rest <= fmax))[0]
    
    # Average PSD across subjects
    avg_psd_rest_band = np.mean(avg_psd_rest[:, idx_band], axis=0)
    avg_psd_task_band = np.mean(avg_psd_task[:, idx_band], axis=0)
    
    # Plotting
    plt.figure(figsize=(10, 6))
    plt.title(f'Average PSD ({band} band)')
    plt.plot(freqs_rest[idx_band], 10 * np.log10(avg_psd_rest_band), label='Rest')  # Convert to dB/Hz
    plt.plot(freqs_task[idx_band], 10 * np.log10(avg_psd_task_band), label='Task')  # Convert to dB/Hz
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Power Spectral Density (dB/Hz)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import numpy as np

def add_gaussian_noise(data, mean=0.0, std=0.01):
    noise = np.random.normal(mean, std, data.shape)
    return data + noise

def time_shift(data, shift):
    return np.roll(data, shift, axis=-1)

def frequency_scaling(data, scale):
    return data * scale

def random_erasing(data, erase_prob=0.1):
    mask = np.random.rand(*data.shape) > erase_prob
    return data * mask

def augment_data(data):
    data = add_gaussian_noise(data)
    data = time_shift(data, shift=np.random.randint(1, 10))
    data = frequency_scaling(data, scale=np.random.uniform(0.9, 1.1))
    data = random_erasing(data, erase_prob=0.1)
    return data
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Model

def ATCNet(input_shape, num_classes=2):
    inputs = tf.keras.Input(shape=input_shape)
    
    # Temporal Convolutional Layers
    x = layers.Conv2D(32, (1, 3), padding='same', activation='relu')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D(pool_size=(1, 2))(x)
    
    x = layers.Conv2D(64, (1, 3), padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D(pool_size=(1, 2))(x)
    
    x = layers.Conv2D(128, (1, 3), padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D(pool_size=(1, 2))(x)
    
    # Fully Connected Layers
    x = layers.Flatten()(x)
    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dropout(0.5)(x)
    
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    
    model = Model(inputs, outputs)
    return model

# Define the input shape and number of classes
input_shape = (42, 51, 1)
num_classes = 2

# Create the model
atcnet_model1 = ATCNet(input_shape, num_classes)

# Compile the model
atcnet_model1.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001), 
                     loss='sparse_categorical_crossentropy', 
                     metrics=['accuracy'])

# Summary of the model
atcnet_model1.summary()
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Model

class Sampling(layers.Layer):
    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = tf.keras.backend.random_normal(shape=(batch, dim))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon

def VAE(input_shape, latent_dim=2):
    # Encoder
    encoder_inputs = tf.keras.Input(shape=input_shape)
    x = layers.Conv2D(32, (3, 3), activation='relu', strides=2, padding='same')(encoder_inputs)
    x = layers.Conv2D(64, (3, 3), activation='relu', strides=2, padding='same')(x)
    x = layers.Flatten()(x)
    x = layers.Dense(16, activation='relu')(x)
    z_mean = layers.Dense(latent_dim, name='z_mean')(x)
    z_log_var = layers.Dense(latent_dim, name='z_log_var')(x)
    z = Sampling()([z_mean, z_log_var])
    encoder = Model(encoder_inputs, [z_mean, z_log_var, z], name='encoder')

    # Decoder
    latent_inputs = tf.keras.Input(shape=(latent_dim,))
    x = layers.Dense(11 * 13 * 64, activation='relu')(latent_inputs)
    x = layers.Reshape((11, 13, 64))(x)
    x = layers.Conv2DTranspose(64, (3, 3), activation='relu', strides=2, padding='same')(x)
    x = layers.Conv2DTranspose(32, (3, 3), activation='relu', strides=2, padding='same')(x)
    decoder_outputs = layers.Conv2DTranspose(1, (3, 3), activation='sigmoid', padding='same')(x)
    decoder = Model(latent_inputs, decoder_outputs, name='decoder')

    # VAE
    vae_outputs = decoder(encoder(encoder_inputs)[2])
    vae = Model(encoder_inputs, vae_outputs, name='vae')

    # Loss function
    reconstruction_loss = tf.keras.losses.binary_crossentropy(tf.keras.backend.flatten(encoder_inputs),
                                                              tf.keras.backend.flatten(vae_outputs))
    reconstruction_loss *= input_shape[0] * input_shape[1]
    kl_loss = 1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var)
    kl_loss = tf.reduce_sum(kl_loss, axis=-1)
    kl_loss *= -0.5
    vae_loss = tf.reduce_mean(reconstruction_loss + kl_loss)
    vae.add_loss(vae_loss)

    return vae, encoder, decoder

# Define the input shape and latent dimension
input_shape = (42, 51, 1)
latent_dim = 2

# Create the VAE model
vae_model, encoder, decoder = VAE(input_shape, latent_dim)

# Compile the VAE model
vae_model.compile(optimizer=tf.keras.optimizers.Adam())

# Summary of the VAE model
vae_model.summary()

# Add a classification head for binary classification
x = layers.Flatten()(encoder.output[2])
x = layers.Dense(128, activation='relu')(x)
x = layers.Dropout(0.5)(x)
outputs = layers.Dense(2, activation='softmax')(x)
classifier = Model(encoder.input, outputs)

# Compile the classifier model
classifier.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001), 
                   loss='sparse_categorical_crossentropy', 
                   metrics=['accuracy'])

# Summary of the classifier model
classifier.summary()
```

```python
import numpy as np

def augment_data(X, y, num_augmentations=9):
    augmented_X = []
    augmented_y = []
    
    for i in range(len(X)):
        for _ in range(num_augmentations):
            # Randomly choose an augmentation technique
            choice = np.random.choice(['noise', 'shift', 'scale'])
            
            if choice == 'noise':
                noise = np.random.normal(0, 0.01, X[i].shape)
                augmented_sample = X[i] + noise
            elif choice == 'shift':
                shift = np.random.randint(1, 10)
                augmented_sample = np.roll(X[i], shift, axis=1)
            elif choice == 'scale':
                scale = np.random.uniform(0.8, 1.2)
                augmented_sample = X[i] * scale
            
            augmented_X.append(augmented_sample)
            augmented_y.append(y[i])
    
    augmented_X = np.array(augmented_X)
    augmented_y = np.array(augmented_y)
    
    # Combine original and augmented data
    X_augmented = np.concatenate((X, augmented_X), axis=0)
    y_augmented = np.concatenate((y, augmented_y), axis=0)
    
    return X_augmented, y_augmented

# Augment the training data by a factor of 10
X_train_augmented, y_train_augmented = augment_data(X_train, y_train, num_augmentations=9)

print("Augmented X_train shape:", X_train_augmented.shape)
print("Augmented y_train shape:", y_train_augmented.shape)
```
